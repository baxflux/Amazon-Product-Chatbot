# **First Dataset - Consumer Reviews of Amazon Products**

```python
first_path = '/kaggle/input/consumer-reviews-of-amazon-products/1429_1.csv'

amazon_df = pd.read_csv(first_path, low_memory=False)

amazon_df.head()
```

```python
amazon_df.info()
amazon_df.isna().sum()
```

```python
print("First 5 Category Strings:")
for i, cat in enumerate(amazon_df['categories'].iloc[:5], start=1):
    print(f"{i}. {cat}")
```

```python
filtered_amazon_df = amazon_df[['asins', 'name', 'categories', 'reviews.text', 'reviews.rating']]
filtered_amazon_df = filtered_amazon_df.dropna().drop_duplicates()
```

```python
filtered_amazon_df['categories'] = filtered_amazon_df['categories'].astype(str).str.split(',').str[0].str.strip()

category_counts = filtered_amazon_df['categories'].value_counts()
category_counts.index.name = "Categories:"

print(category_counts)

top_10_categories = category_counts.head(10).index.tolist()

print("\nTop 10 Categories:")
for i, cat in enumerate(top_10_categories, start=1):
    print(f"{i}. {cat}")
```

```python
filtered_amazon_df = filtered_amazon_df[filtered_amazon_df['categories'].isin(top_10_categories)]

filtered_amazon_df.head()
```

```python
print("Cleaned First Dataset (Top 10 Categories)!")

filtered_amazon_df.to_csv('cleaned_amazon_reviews.csv', index=False)
```

# **Second Dataset - Bitext Gen AI Chatbot Customer Support Dataset**

```python
second_path = '/kaggle/input/bitext-gen-ai-chatbot-customer-support-dataset/Bitext_Sample_Customer_Support_Training_Dataset_27K_responses-v11.csv'

bitext_df = pd.read_csv(second_path, low_memory=False)

bitext_df.head()
```

```python
bitext_df.info()
bitext_df.isna().sum()
```

```python
print("Categories:")
for c in bitext_df['category'].unique():
    print("-", c)

print("\nIntents:")
for i in bitext_df['intent'].unique():
    print("-", i)
```

```python
print("First 5 Instruction Strings:")
for i, cat in enumerate(bitext_df['instruction'].iloc[:5], start=1):
    print(f"{i}. {cat}")
```

```python
print("First 5 Response Strings:")
for i, cat in enumerate(bitext_df['response'].iloc[:5], start=1):
    print(f"\n {i}.\n {cat}")
```

```python
pattern = r'{{[^}]+}}'

placeholders = set()

for col in ['response', 'instruction']:
    for text in bitext_df[col].dropna():
        placeholders.update(re.findall(pattern, text))

print("Unique Placeholders:")
for p in sorted(placeholders):
    print("-", p)
```

```python
filtered_bitext_df = bitext_df[['instruction', 'intent', 'response']]
filtered_bitext_df = filtered_bitext_df.dropna().drop_duplicates()
```

```python
filtered_bitext_df['instruction'] = filtered_bitext_df['instruction'].astype(str)
filtered_bitext_df['response'] = filtered_bitext_df['response'].astype(str)

filtered_bitext_df.head()
```

```python
print("Cleaned Second Dataset!")

filtered_amazon_df.to_csv('cleaned_bitext_intents.csv', index=False)
```
